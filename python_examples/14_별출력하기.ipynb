```python
# 사각형별
for i in range(1,6):
        print('*'*5)

```

```python
# 계단식별
for i in range(1,6):
    print('*'*i)
```

```python
# 대각선별
for i in range(1,6):
    print(' '*(i-1), end='')
    print('*')
```

```python
# 계단식(역순)
for i in range(5,0,-1):
    print('*'*i)
```

```python
# 계단식(역순)
for i in range(5,0,-1):
    print(' '*(5-i),end='')
    print('*'*i)
```
