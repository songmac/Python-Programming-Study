2. 성적관리 프로그램 개발


```python
import pickle
import os

filename = "score_management.p"

def save_data(data, filename):
    with open(filename, "wb") as file:
        pickle.dump(data, file)

def load_data(filename):
    if os.path.exists(filename):
        with open(filename, "rb") as file:
            return pickle.load(file)
    else:
        return []

# 메인 데이터 리스트
score_list = load_data(filename)

menu = int(input("메뉴를 선택해주세요 1-입력, 2-조회, 3-삭제, 0-종료: "))

if menu == 1:
    name = input("이름 : ")
    math = int(input("수학 : "))
    science = int(input("과학 : "))
    english = int(input("영어 : "))
    score_list.append({"이름": name, "수학": math, "과학": science, "영어": english})
    save_data(score_list, filename)

elif menu == 2:
    # 여기서 enumerate()의 세 번째 인자인 1은 인덱스 시작 번호를 1로 설정 # 이렇게 하면 사용자가 리스트의 첫 번째 요소를 '1'로 인식
    for idx, score in enumerate(score_list, 1):
        # idx는 학생의 번호(1부터 시작), score는 이름, 수학, 과학, 영어 점수가 저장된 딕셔너리
        # 딕셔너리에서 각 키('이름', '수학', '과학', '영어')에 해당하는 값을 가져와 출력
        print(f"[{idx}] 이름: {score['이름']}, 수학: {score['수학']}, 과학: {score['과학']}, 영어: {score['영어']}")

elif menu == 3:
    del_num = int(input("삭제할 번호를 입력해주세요 : "))
    # del_num이 유효한 범위 내에 있는지 확인 # del_num은 1부터 score_list의 길이까지 유효해야 함
    if 0 < del_num <= len(score_list):
        # 리스트의 인덱스는 0부터 시작하기 때문에 del_num에서 1을 빼줌
        del score_list[del_num - 1]
        save_data(score_list, filename)
    else:
        print("잘못된 번호입니다.")
    print("삭제가 완료되었습니다.")

elif menu == 0:
    print("종료되었습니다.")

else:
    pass
```
