```python
# 수정 후 : 인스턴스 속성을 수정

class Character:
    def __init__(self):
        self.health = 200

    def move(self):
        self.health -= 10

    def rest(self):
        self.health += 10

    def check_health(self):
        print(f"Health: {self.health}")


class Knight(Character): # 캐릭터1
    def move(self):
        super().move()  # Character 클래스의 move 메서드를 호출하여 health 감소
        self.health -= 5  # Knight 클래스는 추가로 health가 더 감소

    def attack(self):
        print("공격합니다")


class Healer(Character): # 캐릭터2
    def __init__(self):
        super().__init__() # 클래스 상속을 나타냄
        self.mana = 100

    def heal(self, character):
        if self.mana >= 10: # 마나가 충분할 때만 치유 가능
            self.mana -= 10
            character.rest()  # 다른 캐릭터를 치유할 때는 해당 캐릭터의 rest 메서드를 호출
        else:
            print("마나가 부족합니다")

    def check_mana(self):
        print(f"Mana: {self.mana}")


# 테스트
knight = Knight()
healer = Healer()

knight.move()
knight.check_health()
knight.attack()

healer.heal(knight)
healer.check_mana()
knight.check_health()
```
