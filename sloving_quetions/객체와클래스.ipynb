```python
#   :   

class Character:
    def __init__(self):
        self.health = 200

    def move(self):
        self.health -= 10

    def rest(self):
        self.health += 10

    def check_health(self):
        print(f"Health: {self.health}")


class Knight(Character): # 1
    def move(self):
        super().move()  # Character  move   health 
        self.health -= 5  # Knight   health  

    def attack(self):
        print("")


class Healer(Character): # 2
    def __init__(self):
        super().__init__() #   
        self.mana = 100

    def heal(self, character):
        if self.mana >= 10: #     
            self.mana -= 10
            character.rest()  #       rest  
        else:
            print(" ")

    def check_mana(self):
        print(f"Mana: {self.mana}")


# 
knight = Knight()
healer = Healer()

knight.move()
knight.check_health()
knight.attack()

healer.heal(knight)
healer.check_mana()
knight.check_health()
```
