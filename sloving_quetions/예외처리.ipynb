```python
# pickle 모듈은 객체를 파일로 직렬화하거나 파일에서 객체를 역직렬화할 때, 
# 객체 하나당 pickle.dump 또는 pickle.load를 한 번씩 호출

import pickle

try:
    with open("data3.p", "rb") as file:
        data3 = pickle.load(file)
        print(data3)

except FileNotFoundError:  # 파일이 없을 경우에만 새로 쓰도록 예외를 구체화
    with open("data3.p", "wb") as file:
        name = "tom"
        address = "서울시 마포구"
        email = "abcde@gmail.com"
        data_to_save = (name, address, email)  # 데이터를 튜플로 묶음
        pickle.dump(data_to_save, file)  # 한 번의 모든 데이터를 저장
                                        

# pickle.dump 함수 추가 설명
"""
파이썬 객체를 바이트 스트림으로 변환하여 파일에 저장하는데 사용하는데 이 과정을 '직렬화' 또는 '피클링'이라고 함
직렬화란, 메모리에 있는 객체의 상태를 파일이나 네트워크를 통해 전송할 수 있는 형태로 변환하는 것을 말함
이렇게 변환된 바이트 스트림은 나중에 다시 객체로 복원(역직렬화 또는 언피클링)할 수 있음

pickle.dump 함수는 (직렬화하려는 파이썬 객체, 직렬화된 데이터를 쓸 파일 객체)를 인자로 받음
그리고 pickle.dump 함수의 반환값은 None이라는 것은, 이 함수가 호출된 후에 반환되는 값이 없다는 뜻임
즉, 함수는 단지 작업을 수행할 뿐, 호출자에게 어떤 값을 돌려주지 않음
"""
```
