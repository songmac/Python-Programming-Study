```python
#   

class Car:
    car_count = 0

    def __init__(self, name, emission, year):
        self.name = name
        self.emission = emission
        self.year = year
        Car.car_count += 1

    def name_Check(self):
        print(": ", self.name)

    def name_Change(self, new_name):
        self.name = new_name #      print 
        print(" : ", self.name)

    def car_Size(self):
        if self.emission < 1000 :
            print(": ")
        elif self.emission <= 2000 :
            print(": ")
        else:
            print(": ")
    
    @classmethod
    def car_Count(cls): #  self  cls 
        print("  : ", cls.car_count)


aespa = Car("aespa", 2500, 2020)
aespa.name_Check()
aespa.car_Size()
Car.car_Count() #   
aespa.name_Change("aespatwins")
aespa.name_Check()

print()

ive = Car("ive", 900, 2021)
ive.name_Check()
ive.car_Size()
Car.car_Count() 
ive.name_Change("ivetwins")
ive.name_Check()
```
