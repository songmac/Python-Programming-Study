```python
class Car:
    car_count = 0

    def __init__(self, name, emission, year):
        self.name = name
        self.emission = emission
        self.year = year
        Car.car_count += 1

    def name_Check(self):
        print("차이름: ", self.name)

    def name_Change(self, new_name):
        print("변경할 차이름: ", new_name)
        self.name = new_name

    def car_Size(self):
        if self.emission < 1000 :
            print("크기: 소형")
        elif 1000 <= self.emission <= 2000 :
            print("크기: 중형")
        elif 2000 < self.emission :
            print("크기: 대형")
    
    def car_Count(self):
        print("등록된 차량 갯수: ", Car.car_count)



    
aespa = Car("aespa", 2500, 2020)

aespa.name_Check()
aespa.car_Size()
aespa.car_Count()
aespa.name_Change("aespatwins")
aespa.name_Check()

print()

ive = Car("ive", 900, 2021)

ive.name_Check()
ive.car_Size()
ive.car_Count()
ive.name_Change("ivetwins")
ive.name_Check()
```

```python
# 몇가지 수정한 버전

class Car:
    car_count = 0

    def __init__(self, name, emission, year):
        self.name = name
        self.emission = emission
        self.year = year
        Car.car_count += 1

    def name_Check(self):
        print("차이름: ", self.name)

    def name_Change(self, new_name):
        self.name = new_name # 이름을 변경 시킨 후 아래 print문으로 출력
        print("변경된 차이름: ", self.name)

    def car_Size(self):
        if self.emission < 1000 :
            print("크기: 소형")
        elif self.emission <= 2000 :
            print("크기: 중형")
        else:
            print("크기: 대형")
    
    @classmethod
    def car_Count(cls): # 클래스는 self가 아닌 cls로 호출
        print("등록된 차량 갯수: ", cls.car_count)


aespa = Car("aespa", 2500, 2020)
aespa.name_Check()
aespa.car_Size()
Car.car_Count() # 클래스 이름으로 호출
aespa.name_Change("aespatwins")
aespa.name_Check()

print()

ive = Car("ive", 900, 2021)
ive.name_Check()
ive.car_Size()
Car.car_Count() 
ive.name_Change("ivetwins")
ive.name_Check()
```
